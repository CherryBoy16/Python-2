```python
#Task-1
import matplotlib.pyplot as plt
data=[5,6,6,8,7,8,9,10,9,9,11,10]
plt.hist(data,bins=10,edgecolor='black',color='red')
plt.title('Simple Histograph Example')
plt.xlabel('Number')
plt.ylabel('Count')
plt.show()
```

```python
#Task-2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
datf=pd.DataFrame({"Season 1":[6,6,4,4,8,2],
                   "Season 2":[2,4,3,6,8,9]})
p=sns.histplot(data=datf)
p.set(xlabel="X Label Values",ylabel="Y Label Values")
plt.show()
```

```python
#Task-3
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data=np.random.randint(19,71,1000)
plt.hist(data,bins=15,edgecolor='black',color='green')
plt.title('Histogram of Cancer Patients Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Pateints')
plt.show()
```
