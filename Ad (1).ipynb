```python
def smart_dec(func):
    def inner(a, b):
        if b == 0:
            return "Error: cannot divide."
        if a > b:
            a, b = b, a
        return func(a, b)
    return inner
@smart_dec
def div(a, b):
    return a / b
print(div(6, 0))
print(div(6, 2))
```

```python
def smart_dec(func):
    def inner(a, b):
        if a * b > 20:
            return f"Multiplication is greater than 20, performing division: {a / b if b != 0 else 'Error: cannot divide by zero'}"
        return func(a, b)
    return inner
@smart_dec
def mul(a, b):
    return a * b
print(mul(5, 4))
print(mul(2, 3))
```
